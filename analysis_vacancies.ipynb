```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df = pd.read_csv("vacancy.csv")
```

```python
all_technologies = []
for techs in df["technologies"].dropna():
    all_technologies.extend(techs.split(","))
    
tech_counter = Counter(all_technologies)
```

```python
tech_df = pd.DataFrame(tech_counter.items(), columns=["Technology", "Frequency"])
tech_df = tech_df.sort_values(by="Frequency", ascending=False)

sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

colors = sns.color_palette("Blues_r", len(tech_df))
bars = plt.bar(tech_df["Technology"], tech_df["Frequency"], color=colors)

for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval + 0.5,
        int(yval),
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold",
        color="black",
    )

plt.xlabel("Technologies", fontsize=14, fontweight="bold")
plt.ylabel("Frequency", fontsize=14, fontweight="bold")
plt.title("Frequency of Technologies in Job Postings", fontsize=16, fontweight="bold")
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
